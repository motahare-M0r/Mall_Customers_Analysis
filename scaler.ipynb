```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
mall = pd.read_csv("./Mall_Customers.csv")
print(mall)
```

```python
print(mall.info())
```

```python
mall.isna().sum()
```

<h1>Pearsonr</h1>

```python

corr1 = pearsonr(mall["Age"] , mall["Spending Score (1-100)"])


plt.figure(figsize=(8, 5))
plt.scatter(mall["Age"], mall["Spending Score (1-100)"], alpha=0.4 , color='green')
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.title("Correlation between age and Spending Score ")
plt.grid(alpha=0.3)
plt.show()


print(f"Correlation coefficient : {corr1}")
```

```python

corr2 = pearsonr(mall["Age"], mall["Annual Income (k$)"])


plt.figure(figsize=(8, 5))
plt.scatter(mall["Age"], mall["Annual Income (k$)"], alpha=0.4 , color='blue')
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.title("Correlation between age and income")
plt.grid(alpha=0.3)
plt.show()


print(f"Correlation coefficient : {corr2}")
```

```python

corr3 = pearsonr(mall["Spending Score (1-100)"] , mall["Annual Income (k$)"])


plt.figure(figsize=(8, 5))
plt.scatter(mall["Spending Score (1-100)"], mall["Annual Income (k$)"], alpha=0.4 , color='red')
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Annual Income (k$)")
plt.title("Correlation between Annual Income (k$) and Spending Score (1-100)")
plt.grid(alpha=0.3)
plt.show()


print(f"Correlation coefficient : {corr3}")
```

```python
sns.boxplot(data=mall, x="Genre", y="Annual Income (k$)")
plt.title('Income distribution by gender')
plt.show()
```

<h1>Correlatin</h1>

```python
mallist = mall[["CustomerID" , "Age" , "Annual Income (k$)" , "Spending Score (1-100)"]]

corr = mallist.corr()
corr
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(corr , annot=True , fmt="0.2f" , cmap="coolwarm")
```
